```python
import pywt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
root = Path(r"C:\AIEEE\DATASET\Processed_CardRF\Train")
csv_files = list(root.rglob("*.csv"))  # 모든 하위 폴더 포함
print(f"총 {len(csv_files)}개의 CSV 파일이 있습니다.")
```

```python
all_data = []
for f in csv_files:
    # 임시 파일이나 깨진 파일 무시
    if f.name.startswith('~$'):
        continue
    try:
        df = pd.read_csv(f)
        all_data.append(df)
    except Exception as e:
        print(f"⚠️ 오류: {f.name} → {e}")
```
